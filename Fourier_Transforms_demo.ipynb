```python
from scipy import fftpack
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
col_map='viridis'
filenamein1 = 'images/Dog.png'#'Dog.png'#'
filenamein2 = 'images/pintlilColGrey.png'#
img_sky1=mpimg.imread(filenamein1)
img_sky2=mpimg.imread(filenamein2)
```

```python
#=== fft stuff ===#
fft_f1=np.fft.fft2(img_sky1[:,:,1])     #take the 2d FFT of one 'channel' of image 1
fft_sky1 = fftpack.fftshift( fft_f1 )   # shift the zero freq component position to centre of image
fft_f2=np.fft.fft2(img_sky2[:,:,2])     #take the 2d FFT of one 'channel' of image 2         
fft_sky2 = fftpack.fftshift( fft_f2 )   # shift the zero freq component position to centre of image
```

```python
#=== plot sky image
fig_sky = plt.figure(1)
ax_sky = fig_sky.add_subplot(111)
ax_sky.imshow(img_sky1[:,:,1], cmap=col_map)
ax_sky.set_title("sky1")
ax_sky.get_xaxis().set_visible(False)
ax_sky.get_yaxis().set_visible(False)
plt.show()

fig_sky2 = plt.figure(10)
ax_sky2 = fig_sky2.add_subplot(111)
ax_sky2.imshow(img_sky2[:,:,1], cmap=col_map)
ax_sky2.set_title("sky2")
ax_sky2.get_xaxis().set_visible(False)
ax_sky2.get_yaxis().set_visible(False)
plt.show()
```

```python
phase1 = np.angle(fft_sky1)
amp1 = np.abs(fft_sky1)
phase2 = np.angle(fft_sky2)
amp2 = np.abs(fft_sky2)
```

```python
#=== plot phase1 image
fig_phase = plt.figure(4)
ax_phase = fig_phase.add_subplot(111)
ax_phase.imshow(phase1, cmap=col_map)
ax_phase.set_title("phase")
ax_phase.get_xaxis().set_visible(False)
ax_phase.get_yaxis().set_visible(False)
plt.show()

#=== plot amp1 image
fig_amp = plt.figure(5)
ax_amp = fig_amp.add_subplot(111)
ax_amp.imshow(amp1, cmap=col_map, vmin=0, vmax = 1000.0)
ax_amp.set_title("amp")
ax_amp.get_xaxis().set_visible(False)
ax_amp.get_yaxis().set_visible(False)
plt.show()
```

```python
#=== plot phase2 image
fig_phase2 = plt.figure(6)
ax_phase2 = fig_phase2.add_subplot(111)
ax_phase2.imshow(phase2, cmap=col_map)
ax_phase2.set_title("phase")
ax_phase2.get_xaxis().set_visible(False)
ax_phase2.get_yaxis().set_visible(False)
plt.show()

#=== plot amp2 image
fig_amp2 = plt.figure(7)
ax_amp2 = fig_amp2.add_subplot(111)
ax_amp2.imshow(amp2, cmap=col_map, vmin=0, vmax = 1000.0)
ax_amp2.set_title("amp")
ax_amp2.get_xaxis().set_visible(False)
ax_amp2.get_yaxis().set_visible(False)
plt.show()
```

## Swap amps and phases

```python
#=========
#dog amp beer phase

A1ph2 = np.abs(np.fft.ifft2(np.fft.fftshift(amp1*np.exp(1j*phase2))))
A2ph1 = np.abs(np.fft.ifft2(np.fft.fftshift(amp2*np.exp(1j*phase1))))
#=== plot sky image
fig_A1ph2 = plt.figure(9)
ax_A1ph2 = fig_A1ph2.add_subplot(111)
ax_A1ph2.imshow(A1ph2, cmap=col_map)
ax_A1ph2.set_title("Sky: Amplitude 1, Phase 2")
ax_A1ph2.get_xaxis().set_visible(False)
ax_A1ph2.get_yaxis().set_visible(False)
plt.show()


fig_A2ph1 = plt.figure(14)
ax_A2ph1 = fig_A2ph1.add_subplot(111)
ax_A2ph1.imshow(A2ph1, cmap=col_map)
ax_A2ph1.set_title("Sky: Amplitude 2, Phase 1")
ax_A2ph1.get_xaxis().set_visible(False)
ax_A2ph1.get_yaxis().set_visible(False)
plt.show()
```
